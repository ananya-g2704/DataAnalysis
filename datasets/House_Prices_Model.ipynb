# 1. Load libraries & Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset 
df = pd.read_csv("C:/Users/anany/Downloads/train (1).csv")
df.head()
```

```python
# Basic Information & Overview 
print("Shape of dataset: ",df.shape)
print("\n Data Info:\n" )
df.info()

# Summary 
df.describe(include='all')
```

```python
# Missing values analysis

missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percent': (df.isnull().sum()/ len(df))*100
})
missing.sort_values(by='missing_percent',ascending=False)
```

```python
# 4. Handling Missing Values

# Fill numeric missing values with mean
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())

# Fill categorical missing values with mode
for col in ['Alley', 'MiscFeature', 'Fence', 'FireplaceQu']:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].mode()[0])

# Drop 'PoolQC' column only if it exists
df.drop(columns=['PoolQC'], inplace=True, errors='ignore')

print("Missing values after cleaning:\n", df.isnull().sum().sum())
```

```python
# Missing values analysis

missing = pd.DataFrame({
    'missing_count': df.isnull().sum(),
    'missing_percent': (df.isnull().sum()/ len(df))*100
})
missing.sort_values(by='missing_percent',ascending=False)
```

### Insights

1.Filled important numeric and categorical columns with mean or mode to retain key information.

2.Removed columns with very high missing values (like PoolQC, Alley, Fence) to improve data quality.

3.This ensures the dataset is clean, consistent, and ready for further analysis.

```python
# Duplicates & Data type Check
print("Duplicate rows:", df.duplicated().sum())

# Convert selected object columns to category type
for col in ['MSZoning', 'Street', 'Alley', 'Neighborhood', 'HouseStyle', 
            'SaleCondition', 'SaleType']:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Display data info
df.info()
```

```python
# 6. Univariate Analysis - Numerical Features

# Select numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols].describe()

# Histogram: SalePrice
sns.histplot(df['SalePrice'], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Sale Price")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()

# Boxplot: LotArea
sns.boxplot(x=df['LotArea'], color='lightgreen')
plt.title("Lot Area Distribution with Outliers")
plt.xlabel("Lot Area")
plt.show()
```

```python
# Univariate Analysis - Categorical Features

cat_cols = df.select_dtypes(include='category').columns

for col in cat_cols:
    sns.countplot(x=col, data=df)
    plt.title(f'Count of {col}')
    plt.show()
```

```python
# Bivariate Analysis - Numerical vs Categorical

# Boxplot: Neighborhood vs SalePrice
sns.boxplot(x='Neighborhood', y='SalePrice', data=df)
plt.title("Sale Price by Neighborhood")
plt.show()

# Boxplot: Overall Quality vs SalePrice
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title("Sale Price by Overall Quality")
plt.show()
```

```python
# Bivariate Analysis - Categorical vs Categorical

# HouseStyle vs SaleCondition
sns.countplot(x='HouseStyle', hue='SaleCondition', data=df)
plt.title("House Style vs Sale Condition")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# MSZoning vs SaleType
sns.countplot(x='MSZoning', hue='SaleType', data=df)
plt.title("MS Zoning vs Sale Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Correlation and Multivariate Analysis

# Select key numeric columns
corr = df[['OverallQual', 'GrLivArea', 'GarageCars']].corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap - House Prices Dataset")
plt.show()

# Pairplot
sns.pairplot(df[['OverallQual', 'GrLivArea']])
plt.show()
```

#### Insights

1.OverallQual and GrLivArea show a strong positive correlation with SalePrice — meaning better overall quality and larger living areas lead to higher house prices.

2.GarageCars and TotalBsmtSF also have a moderate positive correlation with SalePrice, showing that larger basements and more garage space add value.

3.There are no strong negative correlations, indicating most features contribute positively to property value.

```python
# Outlier Detection using IQR (Example: SalePrice)
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['SalePrice'] < lower) | (df['SalePrice'] > upper)]
print("Number of outliers in SalePrice:", len(outliers))

# Cap outliers
df['SalePrice_capped'] = df['SalePrice'].clip(lower, upper)

# Boxplot after capping
sns.boxplot(x=df['SalePrice_capped'], color='lightblue')
plt.title("SalePrice after Outlier Capping")
plt.xlabel("SalePrice")
plt.show()
```

```python
# Feature Engineering

# Total square footage
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']

# Age of the house
df['HouseAge'] = df['YrSold'] - df['YearBuilt']

# Has garage
df['HasGarage'] = (df['GarageCars'] > 0).astype(int)

# Visualize HasGarage vs SalePrice (binned)
sns.boxplot(x='HasGarage', y='SalePrice', data=df)
plt.title("SalePrice by Garage Availability")
plt.show()

# Display first few rows of new features
df[['TotalSF', 'HouseAge', 'HasGarage', 'SalePrice']].head()
```

```python
# Export the Cleaned House Prices data
df.to_csv("HousePrices_cleaned.csv", index=False)
print("\nCleaned House Prices data saved to HousePrices_cleaned.csv")
```
