# Matplotlib-Python Visualization Library

## What is Matplotlib?

Matplotlib is a python library for creating static,animated and interactive visualizations.

## Why we need it

It helps to understand trends,patterns nad outliers.

## Where we use it?


1. Data Anlysis

2. Machine Learning(For plotting loss curvers)

3. To analyse the market trends,finance,and scientific research

## 2. Importing Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
 
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Bar Plot

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.bar(categories, values, color='blue')
plt.title("Bar Plot Example")
plt.show()
```

### Scatter Plot

```python
x = [5,7,8,7,2,17,2,9,4,11]
y = [99,86,87,88,100,86,103,87,94,78]
 
plt.scatter(x, y, color='blue', marker='.')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Histogram 

```python
data = [22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]
 
plt.hist(data, bins=5, color='lightgreen', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

###  Customization Plot

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
 
plt.plot(x, y, color='purple', linestyle='--', marker='o', label='Line 1')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

### Sub Plot

```python
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
 
plt.subplot(1,2,1)  # 1 row, 2 columns, plot 1
plt.plot(x, y1, color='blue')
plt.title("Plot 1")
 
plt.subplot(1,2,2)  # 1 row, 2 columns, plot 2
plt.plot(x, y2, color='brown')
plt.title("Plot 2")
 
plt.tight_layout()
plt.show()
 
```

```python
plt.plot(x, y1)
plt.title("Save Plot Example")
plt.savefig("my_plot.png")  # Saves in current directory
plt.show()
```

### Pie Chart

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

### Box Plot

```python
data = [25, 30, 35, 35, 40, 45, 50, 55, 60]
plt.boxplot(data)
plt.title("Box Plot Example (Statistical Distribution)")
plt.ylabel("Values")
plt.show()
```

### Step Plot

```python
x = [1,2,3,4,5]
y = [2,3,5,6,8]
plt.step(x, y, where='mid', color='brown')
plt.title("Step Plot Example")
plt.show()
 
```

### Area Plot

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.fill_between(x, y, color='lightpink', alpha=0.6)
plt.title("Area Plot Example")
plt.show()
```

### Contour Plot

```python
x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
 
plt.contour(X, Y, Z, cmap='plasma')
plt.title("Contour Plot Example")
plt.colorbar()
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
plt.title("Filled Contour Plot")
plt.colorbar()
plt.show()
```

### Polar Plot

```python
theta = np.linspace(0, 2*np.pi, 100)
r = np.abs(np.sin(2*theta))
plt.polar(theta, r)
plt.title("Polar Plot Example")
plt.show()
```

### Error Bar Plot

```python
x = np.arange(1,6)
y = np.random.randint(5,15,5)
errors = np.random.rand(5)
 
plt.errorbar(x, y, yerr=errors, fmt='o', ecolor='red', capsize=5)
plt.title("Error Bar Plot Example")
plt.show()
```

### Horizontal Bar Plot

```python
categories = ['A','B','C','D']
values = [10,15,7,12]
 
plt.barh(categories, values, color='purple')
plt.title("Horizontal Bar Plot")
plt.show()
```

### Stack Plot

```python
days = [1,2,3,4,5]
eating = [2,3,4,3,4]
sleeping = [8,7,8,6,7]
working = [8,9,7,8,8]
 
plt.stackplot(days, eating, sleeping, working, labels=['Eating','Sleeping','Working'])
plt.legend()
plt.title("Stack Plot Example")
plt.show()
```

### Hexbin Plot

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hexbin(x, y, gridsize=25, cmap='Blues')
plt.colorbar()
plt.title("Hexbin Plot Example")
plt.show()
```

### 3D Surface Plot

```python
from mpl_toolkits.mplot3d import Axes3D
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_title("3D Surface Plot")
plt.show()
```

### 3D Line Plot

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t
 
ax.plot3D(x, y, z, 'purple')
ax.set_title("3D Line Plot")
plt.show()
```

### 3D Scatter Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
 
ax.scatter(x, y, z, color='red', marker='o')
ax.set_title("3D Scatter Plot")
plt.show()
```

### 3D Wireframe Plot

```python
x = np.linspace(-5, 5, 30)
y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(X, Y, Z, color='green')
ax.set_title("3D Wireframe Plot")
plt.show()
```

### 3D Contour Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour3D(X, Y, Z, 50, cmap='plasma')
ax.set_title("3D Contour Plot")
plt.show()
```

### 3D Bar Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]
z = np.zeros(5)
dx = np.ones(5)
dy = np.ones(5)
dz = [5, 2, 7, 3, 6]
ax.bar3d(x, y, z, dx, dy, dz, color='teal')
ax.set_title("3D Bar Plot")
plt.show()
```

### 3D Surface with Colormap

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.cos(np.sqrt(X**2 + Y**2))
 
surf = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')
fig.colorbar(surf)
ax.set_title("3D Surface with Colormap")
plt.show()
```

### 3D Sphere Surface

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='skyblue')
ax.set_title("3D Sphere Surface")
plt.show()
```

### 3D Surface + Wireframe

```python
X = np.linspace(-5, 5, 30)
Y = np.linspace(-5, 5, 30)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')
ax.plot_wireframe(X, Y, Z, color='black', linewidth=0.5)
ax.set_title("3D Surface + Wireframe")
plt.show()
```

### 3D Contour Projection on Planes

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
 
ax.contour(X, Y, Z, zdir='z', offset=-2, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=-5, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=5, cmap='coolwarm')
 
ax.set_title("3D Contour Projection on Planes")
plt.show()
```

### Hands-On Excercise

#### 1.Create a line plot of student marks across 5 exams.

```python
exams = ['Exam 1', 'Exam 2', 'Exam 3', 'Exam 4', 'Exam 5']
marks = [78, 85, 69, 92, 88]
plt.plot(exams, marks, color='blue')
plt.title('Student Marks Across 5 Exams')
plt.xlabel('Exams')
plt.ylabel('Marks')
plt.show()
```

#### 2.Plot a bar chart of sales of 4 products.

```python
products = ['A', 'B', 'C', 'D']
sales = [250, 400, 150, 300]
plt.bar(products, sales, color="green")
plt.title('Sales of 4 Products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

#### 3.Make a scatter plot of Age vs Salary.

```python
age = [22, 25, 27, 35, 45, 52, 23, 34, 41, 29]
salary = [25000, 27000, 30000, 40000, 52000, 58000, 26000, 39000, 50000, 33000]
plt.scatter(age, salary, color='red', marker='+')
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

#### 4.Draw a histogram of marks scored by students in a class.

```python
marks_class = [45, 56, 67, 89, 90, 76, 88, 92, 53, 60, 72, 81, 95, 77, 68]
plt.hist(marks_class, bins=5, color='Teal', edgecolor='black')
plt.title('Histogram of Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.show()
```

#### 5.Customize a plot with different colors, markers, and line styles.

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]
plt.plot(x, y, color='green', marker='o', linestyle='--', label='Student A')
plt.title('Customized Line Plot Example')
plt.xlabel('Exam Number')
plt.ylabel('Marks')
plt.legend()
plt.show()
```

#### 6.Create two subplots: Line plot and bar plot side by side.

```python
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
plt.subplot(1,2,1)  
plt.plot(x, y1, color='Turquoise')
plt.title("Plot 1")
plt.subplot(1,2,2)  
plt.bar(x, y2, color='Coral')
plt.title("Plot 2")
plt.tight_layout()
plt.show()
```
